```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV into a DataFrame
df = pd.read_csv('SD1.csv')

# Remove the date and keep hours, minutes, seconds, and milliseconds
df['time'] = df['timestamp'].apply(lambda x: ':'.join(x.split(' ')[1].split(':')[:4]))

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['time'], df['ph'], label='pH')
plt.plot(df['time'], df['salinity'], label='Salinity')
plt.plot(df['time'], df['moisture'], label='Moisture')
plt.plot(df['time'], df['tds'], label='TDS')
plt.plot(df['time'], df['light'], label='Light')
plt.plot(df['time'], df['temperature'], label='Temperature')
plt.plot(df['time'], df['humidity'], label='Humidity')

# Customize the plot
plt.xlabel('Time (HH:MM:SS.sss)')
plt.ylabel('Values')
plt.title('Sensor Data over Time-Line Graph')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV into a DataFrame
df = pd.read_csv('SD1.csv')

# Remove the date and keep hours, minutes, seconds, and milliseconds
df['time'] = df['timestamp'].apply(lambda x: ':'.join(x.split(' ')[1].split(':')[:4]))

# Create subplots
fig, axs = plt.subplots(3, 2, figsize=(15, 15))

# Plot each sensor's data
axs[0, 0].plot(df['time'], df['ph'], label='pH')
axs[0, 0].set_title('pH over Time')
axs[0, 1].plot(df['time'], df['salinity'], label='Salinity')
axs[0, 1].set_title('Salinity over Time')
axs[1, 0].plot(df['time'], df['moisture'], label='Moisture')
axs[1, 0].set_title('Moisture over Time')
axs[1, 1].plot(df['time'], df['tds'], label='TDS')
axs[1, 1].set_title('TDS over Time')
axs[2, 0].plot(df['time'], df['light'], label='Light')
axs[2, 0].set_title('Light over Time')
axs[2, 1].plot(df['time'], df['temperature'], label='Temperature')
axs[2, 1].set_title('Temperature over Time')

# Customize the plot
for ax in axs.flat:
    ax.set(xlabel='Time (HH:MM:SS.sss)', ylabel='Values')
    ax.legend()
    ax.grid(True)
    ax.tick_params(axis='x', rotation=45)

# Hide the empty subplots
axs[2, 1].axis('off')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV into a DataFrame
df = pd.read_csv('SD1.csv')

# Remove the date and keep hours, minutes, seconds, and milliseconds
df['time'] = df['timestamp'].apply(lambda x: ':'.join(x.split(' ')[1].split(':')[:4]))

# Plot each numerical column against timestamp with adjusted marker size and transparency
numerical_columns = ['ph', 'salinity', 'moisture', 'tds', 'light', 'temperature', 'humidity']

plt.figure(figsize=(12, 8))

for col in numerical_columns:
    plt.scatter(df['time'], df[col], label=f'{col} vs Time', alpha=0.7, s=20)

plt.xlabel('Time (HH:MM:SS.sss)')
plt.ylabel('Values')
plt.title('Sensor Data vs Time')
plt.legend()
# Remove grid lines
# plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV into a DataFrame
df = pd.read_csv('SD1.csv')

# Remove the date and keep hours, minutes, seconds, and milliseconds
df['time'] = df['timestamp'].apply(lambda x: ':'.join(x.split(' ')[1].split(':')[:4]))

# Set the style to remove gray background
plt.style.use('default')

# Plot each numerical column against timestamp using a bar graph
numerical_columns = ['ph', 'salinity', 'moisture', 'tds', 'light', 'temperature', 'humidity']

plt.figure(figsize=(12, 8))

for col in numerical_columns:
    plt.bar(df['time'], df[col], label=f'{col} vs Time', alpha=0.7)

plt.xlabel('Time (HH:MM:SS.sss)')
plt.ylabel('Values')
plt.title('Sensor Data vs Time (Bar Graph)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
